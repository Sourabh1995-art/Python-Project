```python
import pandas as pd
df1 = pd.read_csv('customer_policy_details.csv')
df1
```

```python
df1.columns = ['Customer ID','Annual Premium','Sales Channel Code','Vintage', 'Response']
df1
```

```python
#count of null values

#'Count of Null Values Columnwise: ',
df1.isna().sum()
```

```python
#Dropping Null Values
df1.dropna(subset = ['Customer ID'], inplace = True)
```

```python
df1.isna().sum()
```

```python
#Replacing with Mean Values
df1['Annual Premium'].fillna(df1['Annual Premium'].mean(),inplace = True)
```

```python
#Replacing with mode Value
df1['Vintage'].fillna(df1['Vintage'].mode()[0],inplace=True)
df1['Response'].fillna(df1['Response'].mode()[0],inplace=True)
df1['Sales Channel Code'].fillna(df1['Sales Channel Code'].mode()[0],inplace=True)
```

```python
df1.isna().sum()
```

```python
#Outliers
q1 = df1.quantile(0.25)
q3 = df1.quantile(0.75)
```

```python
IQR1 = q3 - q1
```

```python
col1= ((df1<(q1-1.5*IQR1)) | (df1>(q3+1.5*IQR1))).sum()
print(col1)
```

```python
#Merger of 2 columns
merge_df = pd.merge(df,df1,on = 'Customer ID')
merge_df
```

```python
q1 = df1[["Annual Premium"]].quantile(0.25)
q3 = df1[["Annual Premium"]].quantile(0.75)

IQR2 = q3 - q1
```

```python
col1= ((df1[["Annual Premium"]]<(q1-1.5*IQR2)) | (df1[["Annual Premium"]]>(q3+1.5*IQR2))).sum()
print(col1)
```

```python
col1= ((df1[["Annual Premium"]]<(q1-1.5*IQR2)) | (df1[["Annual Premium"]]>(q3+1.5*IQR2)))
ind = col1[col1["Annual Premium"]].index
ind
```

```python
df1.loc[ind,"Annual Premium"] = df1["Annual Premium"].mean()
```

```python
q2 = df1[["Response"]].quantile(0.25)
q4 = df1[["Response"]].quantile(0.75)

IQR3 = q4 - q2
```

```python
col1= ((df1[["Response"]]<(q2-1.5*IQR3)) | (df1[["Response"]]>(q4+1.5*IQR3))).sum()
print(col1)
```

```python
col1= ((df1[["Response"]]<(q2-1.5*IQR3)) | (df1[["Response"]]>(q4+1.5*IQR3)))
ind = col1[col1["Response"]].index
ind
```

```python
df1.loc[ind,"Response"] = df1["Response"].mean()
```
